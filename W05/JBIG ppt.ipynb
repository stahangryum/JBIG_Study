# 보팅(Voting)과 배깅(Bagging)



- 보팅과 배깅은 여러 개의 분류기가 투표를 통해 최종 예측 결과를 결정하는 방식입니다

- 보팅과 배깅의 차이점?

- 보팅은 서로 다른 알고리즘을 가진 분류기 결합

- 배깅은 각각의 분류기가 같은 유형의 알고리즘이지만 데이터 샘플링은 다르게 가져간다!!
