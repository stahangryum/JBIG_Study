앙상블 학습의 유형 

- 보팅(Voting)

- 배깅(Bagging)

- 부스팅(Boosting)


보팅과 배깅은 여러 개의 분류기가 투표를 통해 최종 예측 결과를 결정

보팅은 일반적으로 서로 다른 알고리즘을 가진 분류기가 결합하는 것

배깅의 경우 각각의 분류기가 모두 같은 유형의 알고리즘 기반이지만, 데이터 샘플링을 서로 다르게 가져가면서 학습을 수행해 보팅을 수행하는 것입니다.
